```python
from read_log import parse_log_file
import pandas as pd
from plotly import express as px
```

```python
market_data: pd.DataFrame

sandbox, market_data, trades = parse_log_file('logs/v1.log')
```

```python
trades = pd.DataFrame(trades)
```

```python
own_trades = trades[(trades['buyer'] == 'SUBMISSION') | (trades['seller'] == 'SUBMISSION')]
```

```python
market_data.sort_values(['timestamp', 'product'], inplace=True)

resin = market_data[market_data['product'] == 'RAINFOREST_RESIN']
kelp = market_data[market_data['product'] == 'KELP']
```

```python
# plot resin['profit_and_loss'] vs resin['timestamp']
fig = px.line(resin, x='timestamp', y='profit_and_loss', title='Profit and Loss for Resin')
fig.show()
```

```python
# plot kelp['profit_and_loss'] vs kelp['timestamp']
fig = px.line(kelp, x='timestamp', y='profit_and_loss', title='Profit and Loss for Kelp')
fig.show()
```

```python
# Group by timestamp and sum the profit_and_loss column
aggregated_data = market_data.groupby('timestamp')['profit_and_loss'].sum().reset_index()

# Plot the aggregated profit_and_loss for each timestamp
fig = px.line(aggregated_data, x='timestamp', y='profit_and_loss', title='Aggregated Profit and Loss by Timestamp')
fig.show()
```

```python
print('Resin final profit: ', resin.iloc[-1]['profit_and_loss'])
print('Kelp final profit: ', kelp.iloc[-1]['profit_and_loss'])
print('Total profit: ', aggregated_data.iloc[-1]['profit_and_loss'])
```
