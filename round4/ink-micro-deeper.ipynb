```python
def bollinger_bands(data, column='mid_price', window=20, num_std=2):
    data['SMA'] = data[column].rolling(window=window).mean()
    data['STD'] = data[column].rolling(window=window).std()
    data['UPPER_BAND'] = data['SMA'] + (data['STD'] * num_std)
    data['LOWER_BAND'] = data['SMA'] - (data['STD'] * num_std)
    return data

# browse = bollinger_bands(browse, column='mid_price', window=20, num_std=2)
bollinger_bands(df, column='mid_price', window=100, num_std=2)

plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['mid_price'], label='Mid Price')
plt.plot(df['time'], df['SMA'], label='SMA', color='blue')
plt.plot(df['time'], df['UPPER_BAND'], label='Upper Band', color='red')
plt.plot(df['time'], df['LOWER_BAND'], label='Lower Band', color='green')
plt.fill_between(df['time'], df['UPPER_BAND'], df['LOWER_BAND'], color='lightgray', alpha=0.5)
plt.title('Bollinger Bands')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from tqdm import tqdm
def identify_regime(data, lookback=20):
    regimes = []
    
    for i in range(len(data)):
        if i < lookback:
            regimes.append('NEUTRAL')
        else:
            period = data.iloc[i-lookback+1:i+1]
            price = period['mid_price']
            upper = period['UPPER_BAND']
            lower = period['LOWER_BAND']
            
            # Count contiguous touches of upper and lower bands
            touched_upper = 0
            touched_lower = 0
            touching_upper = False
            touching_lower = False
            
            for j in range(len(period)):
                if price.iloc[j] >= upper.iloc[j]:
                    if not touching_upper:
                        touched_upper += 1
                        touching_upper = True
                    touching_lower = False
                elif price.iloc[j] <= lower.iloc[j]:
                    if not touching_lower:
                        touched_lower += 1
                        touching_lower = True
                    touching_upper = False
                else:
                    touching_upper = False
                    touching_lower = False
            
            if touched_upper > 1 and touched_lower > 1:
                regimes.append('OSCILLATING')
            elif touched_upper > 1:
                regimes.append('RIDING_UPPER')
            elif touched_lower > 1:
                regimes.append('RIDING_LOWER')
            else:
                regimes.append('NEUTRAL')
    
    return pd.Series(regimes, index=data.index)

regimes = identify_regime(data, lookback=300)
```

```python
from collections import deque
import numpy as np
import pandas as pd

# Constants
PRODUCT = 'SQUID_INK'
POSITION_LIMIT = 50
WINDOW_SIZE = 30

# Initialize variables
window = deque()
signal = []
position_history = []
cash = 0
position = 0

# Process data
for idx, row in data.reset_index().iterrows():
    mid = row['mid_price']  # Assuming mid price is available in data
    
    # Initial window filling
    if len(window) < WINDOW_SIZE:
        signal.append('NEUTRAL')
        window.append(mid)
        position_history.append(position)
        continue
    
    # Calculate statistics
    mean = np.mean(window)
    std = np.std(window)
    
    # Calculate z-score (normalized deviation from mean)
    z = (mid - mean) / std if std > 0 else 0
    
    # Trading logic
    take_threshold = 5
    clear_threshold = 1
    
    if z >= take_threshold:
        signal.append('SHORT')
        # Logic to sell would go here
        new_position = max(-POSITION_LIMIT, position - min(position + POSITION_LIMIT, position))
        position = new_position
    elif z <= -take_threshold:
        signal.append('LONG')
        # Logic to buy would go here
        new_position = min(POSITION_LIMIT, position + min(POSITION_LIMIT - position, POSITION_LIMIT))
        position = new_position
    elif -clear_threshold <= z <= clear_threshold:
        if position > 0:
            signal.append('NEUTRAL')
            # Clear long position
            position = 0
        elif position < 0:
            signal.append('NEUTRAL')
            # Clear short position
            position = 0
        else:
            signal.append('NEUTRAL')
    else:
        signal.append('NEUTRAL')
    
    # Update window with new mid price
    window.append(mid)
    if len(window) > WINDOW_SIZE:
        window.popleft()
    
    # Record position
    position_history.append(position)

# Convert to pandas Series
signal = pd.Series(signal, index=data.index)
position_history = pd.Series(position_history, index=data.index)

# You could also compute and store means and standard deviations
means = []
stds = []
for i in range(len(data)):
    if i < WINDOW_SIZE:
        means.append(np.nan)
        stds.append(np.nan)
    else:
        window_slice = list(window)[:i]
        means.append(np.mean(window_slice))
        stds.append(np.std(window_slice))

means = pd.Series(means, index=data.index)
stds = pd.Series(stds, index=data.index)
```

```python
def run_backtest(data, product='SQUID_INK', position_limit=50, window_size=30, take_threshold=5, clear_threshold=1):
    window = deque(maxlen=window_size)
    signal = []
    position_history = []
    position = 0
    cash = 0
    pnl_history = []
    
    for idx, row in data.reset_index().iterrows():
        mid = row['mid_price']
        
        # Calculate previous PnL before position change
        previous_pnl = cash + position * mid if idx > 0 else 0
        
        if len(window) < window_size:
            signal.append('NEUTRAL')
            window.append(mid)
            position_history.append(position)
            pnl_history.append(previous_pnl)
            continue
        
        mean = np.mean(window)
        std = np.std(window)
        z = (mid - mean) / std if std > 1e-8 else 0
        
        previous_position = position
        
        if z >= take_threshold:
            signal.append('SHORT')
            new_position = -position_limit
            position = new_position
        elif z <= -take_threshold:
            signal.append('LONG')
            new_position = position_limit
            position = new_position
        elif -clear_threshold <= z <= clear_threshold:
            if position > 0:
                signal.append('NEUTRAL')
                position = 0
            elif position < 0:
                signal.append('NEUTRAL')
                position = 0
            else:
                signal.append('NEUTRAL')
        else:
            signal.append('NEUTRAL')
        
        # Calculate position change and update cash with cross spread
        position_change = position - previous_position
        if position_change != 0:
            cash -= position_change * mid
            # Apply cross spread fee
            cash = cross_spread(cash, position_change)
        
        # Calculate current PnL
        current_pnl = cash + position * mid
        
        window.append(mid)
        position_history.append(position)
        pnl_history.append(current_pnl)
    
    signal = pd.Series(signal, index=data.index)
    position_history = pd.Series(position_history, index=data.index)
    pnl_series = pd.Series(pnl_history, index=data.index)
    
    max_drawdown = calculate_max_drawdown(pnl_series)
    sharpe_ratio = calculate_sharpe_ratio(pnl_series)
    
    results = {
        'product': product,
        'position_limit': position_limit,
        'window_size': window_size,
        'take_threshold': take_threshold,
        'clear_threshold': clear_threshold,
        'final_pnl': pnl_series.iloc[-1],
        'max_drawdown': max_drawdown,
        'sharpe_ratio': sharpe_ratio,
        'position_history': position_history,
        'pnl_series': pnl_series
    }
    
    return results

def cross_spread(cash, quantity):
    return cash - abs(quantity) * 5

def calculate_max_drawdown(pnl_series):
    running_max = pnl_series.cummax()
    drawdown = (pnl_series - running_max) / running_max.replace(0, 1)  # Avoid division by zero
    max_drawdown = drawdown.min()
    return max_drawdown

def calculate_sharpe_ratio(pnl_series, risk_free_rate=0.0, periods_per_year=252):
    returns = pnl_series.pct_change().dropna()
    excess_returns = returns - risk_free_rate / periods_per_year
    if len(excess_returns) > 0:
        sharpe_ratio = np.sqrt(periods_per_year) * excess_returns.mean() / excess_returns.std()
        return sharpe_ratio
    return np.nan

def run_parameter_sweep(data, products=['SQUID_INK'], position_limits=[30, 50, 100],
                      window_sizes=[20, 30, 40], take_thresholds=[3, 5, 7],
                      clear_thresholds=[0.5, 1, 1.5]):
    results = []
    total_combinations = (len(products) * len(position_limits) * 
                         len(window_sizes) * len(take_thresholds) * 
                         len(clear_thresholds))
    combination_count = 0
    
    for product in products:
        for position_limit in position_limits:
            for window_size in window_sizes:
                for take_threshold in take_thresholds:
                    for clear_threshold in clear_thresholds:
                        combination_count += 1
                        print(f"Running combination {combination_count}/{total_combinations}: "
                             f"product={product}, position_limit={position_limit}, "
                             f"window_size={window_size}, take_threshold={take_threshold}, "
                             f"clear_threshold={clear_threshold}")
                        
                        result = run_backtest(
                            data=data,
                            product=product,
                            position_limit=position_limit,
                            window_size=window_size,
                            take_threshold=take_threshold,
                            clear_threshold=clear_threshold
                        )
                        
                        results.append(result)
    
    return results

def analyze_results(results_list):
    results_df = pd.DataFrame([
        {k: v for k, v in result.items() if k not in ['position_history', 'pnl_series']}
        for result in results_list
    ])
    
    results_df = results_df.sort_values('sharpe_ratio', ascending=False)
    
    print("\nTop 5 strategies by Sharpe ratio:")
    print(results_df.head(5)[['product', 'position_limit', 'window_size', 
                             'take_threshold', 'clear_threshold',
                             'final_pnl', 'max_drawdown', 'sharpe_ratio']])
    
    if len(results_list) > 0:
        top_strategy = results_list[results_df.index[0]]
        
        plt.figure(figsize=(12, 6))
        plt.plot(top_strategy['pnl_series'], label='PnL')
        plt.title(f"Cumulative PnL for Top Strategy\n"
                 f"product={top_strategy['product']}, position_limit={top_strategy['position_limit']}, "
                 f"window_size={top_strategy['window_size']}, take_threshold={top_strategy['take_threshold']}, "
                 f"clear_threshold={top_strategy['clear_threshold']}")
        plt.xlabel('Time')
        plt.ylabel('PnL')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    return results_df
```

```python
# Calculate PnL from trading
def calculate_pnl(data, position_history):
    df = data.copy()

    # Initialize cash column
    df['CASH'] = 0
    
    # Calculate position changes
    df['POSITION_DIFF'] = position_history.diff().fillna(0)
    
    # Calculate cash flow from trades (using the mid price at time of trade)
    df.loc[df['POSITION_DIFF'] != 0, 'CASH'] = -df['POSITION_DIFF'] * df['mid_price']
    
    # Calculate cumulative cash
    df['CUMULATIVE_CASH'] = df['CASH'].cumsum()
    
    # Calculate PnL (cash + market value of current position)
    df['PNL'] = df['CUMULATIVE_CASH'] + position_history * df['mid_price']
    
    return df

# After running the trading algorithm
df_pnl = calculate_pnl(data, position_history)


plt.figure(figsize=(12, 6))
plt.plot(df_pnl.index, df_pnl['PNL'], label='PNL')
plt.title('Cumulative PnL for SQUID_INK Trading')
plt.xlabel('Time')
plt.ylabel('PnL')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
def generate_signals(data):
    price = data['mid_price']
    upper = data['UPPER_BAND']
    lower = data['LOWER_BAND']
    regime = data['REGIME']
    sma = data['SMA']
    
    signals = pd.Series(index=data.index, dtype=object)
    
    for i in tqdm(range(1, len(data))):
        if regime[i] == 'OSCILLATING':
            if price[i] >= upper[i] and price[i-1] < upper[i-1]:
                signals[i] = 'SHORT'
            elif price[i] <= lower[i] and price[i-1] > lower[i-1]:
                signals[i] = 'LONG'
            
        elif regime[i] == 'NEUTRAL' and regime[i-1] != 'NEUTRAL':
            signals[i] = 'CLEAR'
        elif regime[i] == 'RIDING_UPPER' and regime[i-1] != 'RIDING_UPPER':
            if price[i] > sma[i]:
                signals[i] = 'CLEAR'
            else:
                signals[i] = 'LONG'
        elif regime[i] == 'RIDING_LOWER' and regime[i-1] != 'RIDING_LOWER':
            if price[i] < sma[i]:
                signals[i] = 'CLEAR'
            else:
                signals[i] = 'SHORT'
    
    return signals

data['SIGNAL'] = generate_signals(data)
```

```python
def update_position(signals):
    last_signal = None
    positions = []
    
    for signal in signals:
        if pd.notna(signal):
            last_signal = signal
        
        if last_signal == 'LONG':
            position = 60
        elif last_signal == 'SHORT':
            position = -60
        else:
            position = 0
        
        positions.append(position)
    
    return pd.Series(positions, index=signals.index)

data['POSITION'] = update_position(data['SIGNAL'])
```
