```python
# Dictionary of dictionaries representing the exchange rates
exchange_rates = {
     "Snowball": {
        "Snowball": 1,
        "Pizza Slice": 1.45,
        "Silicon Nuggets": 0.52,
        "Shells": 0.72,
    },
    "Pizza Slice": {
        "Snowball": .7,
        "Pizza Slice": 1,
        "Silicon Nuggets": 0.31,
        "Shells": 0.48,
    },
    "Silicon Nuggets": {
        "Snowball": 1.95,
        "Pizza Slice": 3.1,
        "Silicon Nuggets": 1,
        "Shells": 1.49,
    },
    "Shells": {
        "Snowball": 1.34,
        "Pizza Slice": 1.98,
        "Silicon Nuggets": 0.64,
        "Shells": 1,
    }
}
```

```python
# start with 500K Shells
shells = 500000

# trade S -> X -> Y -> S
# where x and y are the other currencies
# we want to find the maximum amount of shells we can get after trading

# we can do this by iterating through all possible trades
# and keeping track of the maximum amount of shells we can get

max_shells = shells
intermediate_currencies = []

all_currencies = exchange_rates.keys()

for currency1 in all_currencies:
    for currency2 in all_currencies:
        if currency1 == currency2:
            continue
        for currency3 in all_currencies:
            if currency1 == currency3 or currency2 == currency3:
                continue
            print(f"shells = {shells}, type = {type(shells)}")
            # print(f"shells = {exchange_rates["Shells"][currency1]}, type = {type(exchange_rates["Shells"][currency1])}")


            # trade S -> X
            shells_to_x = shells * exchange_rates["Shells"][currency1]
            # trade X -> Y
            x_to_y = shells_to_x * exchange_rates[currency1][currency2]
            # trade Y -> S
            y_to_s = x_to_y * exchange_rates[currency2]["Shells"]

            # print the currencies in order and the amount of shells
            print(f"Shells -> {currency1} -> {currency2} -> {currency3} -> Shells: {y_to_s}")

            if y_to_s > max_shells:
                max_shells = y_to_s
                intermediate_currencies = [currency1, currency2, currency3]

print(f"Maximum amount of shells: {max_shells}")
print(f"Intermediate currencies: {intermediate_currencies}")
```

```python
# do it with 5 trades
max_shells = shells
intermediate_currencies = []
max_shells_3 = shells
intermediate_currencies_3 = []

for currency1 in all_currencies:
    for currency2 in all_currencies:
        for currency3 in all_currencies:
             # trade S -> X
            shells_to_a = shells * exchange_rates["Shells"][currency1]
            # trade X -> Y
            a_to_b = shells_to_a * exchange_rates[currency1][currency2]
            # trade Y -> Z
            b_to_c = a_to_b * exchange_rates[currency2][currency3]
            # back to shells
            c_to_s = b_to_c * exchange_rates[currency3]["Shells"]

            # chain and profit
            print(f"Shells -> {currency1} -> {currency2} -> {currency3}\t -> Shells: {c_to_s}")

            if c_to_s > max_shells_3:
                    max_shells_3 = c_to_s
                    intermediate_currencies_3 = [currency1, currency2, currency3]


            for currency4 in all_currencies:

                    # trade S -> X
                    shells_to_x = shells * exchange_rates["Shells"][currency1]
                    # trade X -> Y
                    x_to_y = shells_to_x * exchange_rates[currency1][currency2]
                    # trade Y -> Z
                    y_to_z = x_to_y * exchange_rates[currency2][currency3]
                    # trade Z -> S
                    z_to_w = y_to_z * exchange_rates[currency3][currency4]

                    w_to_s = z_to_w * exchange_rates[currency4]["Shells"]

                    # print shells, shells_to_x, x_to_y, y_to_z, z_to_s
                    # print(shells, shells_to_x, x_to_y, y_to_z, z_to_w, w_to_s)

                    # print the currencies in order and the amount of shells
                    print(f"Shells -> {currency1} -> {currency2} -> {currency3}\t -> {currency4} -> Shells: {w_to_s}")

                    if w_to_s > max_shells:
                        max_shells = w_to_s
                        intermediate_currencies = [currency1, currency2, currency3, currency4]

print(f"Maximum amount of shells: {max_shells}")
print(f"Intermediate currencies: {intermediate_currencies}")

print (f"Maximum amount of shells with 3 intermediary trades: {max_shells_3}")
print(f"Intermediate currencies: {intermediate_currencies_3}")
```

```python
# find the percent return of intermediate_currencies
percent_return = (max_shells - shells) / shells * 100

print(f"Percent return: {percent_return:.2f}%")
```

```python
# find which goods are undervalued relative to the others
# by comparing the exchange rates

lowest = 100
lowest_currency = None
highest = 0
highest_currency = None

for currency1 in all_currencies:
    for currency2 in all_currencies:
        if currency1 == currency2:
            continue
        # print(f"{currency1} -> {currency2}: {exchange_rates[currency1][currency2]} versus {1/exchange_rates[currency2][currency1]}")
        # print the percentage difference
        print(f"{currency1} -> {currency2}: {exchange_rates[currency1][currency2] / (1/exchange_rates[currency2][currency1])}")

        percent_diff = exchange_rates[currency1][currency2] / (1/exchange_rates[currency2][currency1])

        if percent_diff < lowest:
            lowest = percent_diff
            lowest_currency = (currency1, currency2)
        if percent_diff > highest:
            highest = percent_diff
            highest_currency = (currency1, currency2)

print(f"Lowest: {lowest_currency} with {lowest}")
print(f"Highest: {highest_currency} with {highest}")
```
