```python
import orjson
from io import StringIO
import pandas as pd

def read_submission_log(filename: str) -> tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    with open(filename, 'r') as f:
        log_content = f.read()

    sandbox, market = log_content.split("Sandbox logs:")[1].split("Activities log:")
    market, trade_history = market.split("Trade History:")

    market = pd.read_csv(StringIO(market.strip()), sep=';')
    trade_history = orjson.loads(trade_history.strip())
    trade_history = pd.DataFrame(trade_history)
    # sandbox = [orjson.loads(s) for s in re.findall(r'\{.*?\}', sandbox, re.DOTALL)]
    # sandbox = pd.DataFrame(sandbox)
    return market, trade_history, sandbox
```

```python
import pandas as pd
import numpy as np

def compare_dataframes(df1, df2):
    """
    Compare two dataframes including proper handling of NaN values.
    Returns True if dataframes are identical (considering NaN == NaN as True).
    """
    # Check if dataframes have same shape
    if df1.shape != df2.shape:
        print("Dataframes have different shapes")
        return False
    
    # Check if column names match
    if not df1.columns.equals(df2.columns):
        print("Dataframes have different column names")
        return False
    
    # Check if index values match
    if not df1.index.equals(df2.index):
        print("Dataframes have different indices")
        return False
    
    # For each column, check values
    for col in df1.columns:
        # Create masks for NaN values in both dataframes
        mask_nan1 = df1[col].isna()
        mask_nan2 = df2[col].isna()
        
        # Check if NaNs are in the same positions
        if not (mask_nan1 == mask_nan2).all():
            print(f"NaN values don't match in column '{col}'")
            return False
        
        # Compare non-NaN values
        vals1 = df1.loc[~mask_nan1, col]
        vals2 = df2.loc[~mask_nan2, col]
        
        if not vals1.equals(vals2):
            print(f"Values don't match in column '{col}'")
            return False
    
    # All checks passed
    print("Dataframes are identical (NaN values considered equal)")
    return True

# Usage example:
# df1 = pd.read_csv('source1.csv')
# df2 = pd.read_csv('source2.csv')
# are_identical = compare_dataframes(df1, df2)
```
